```python
import tensorflow as tf
print(tf.__version__)
```

```python
# create a data
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 2))
# Create features (using tensors)
X = tf.constant([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

# Create labels (using tensors)
y = tf.constant([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

# Visualize it
plt.scatter(X, y)
```

```python
# random state
tf.random.set_seed(42)

# create a model using the sequential API
model= tf.keras.Sequential([tf.keras.layers.Dense(1)])
```

```python
# See above I have a single Neuron
```

```python
# compile the model
model.compile(loss=tf.keras.losses.mse,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mse"])
```

```python
X
```

```python
#It is in array form, need to change dimension into column
tf.expand_dims(X, axis = -1)
```

```python
# Fit the model
model.fit(tf.expand_dims(X, axis = -1), y, epochs = 20)
```

```python
# Saving all the mse values in a variable

mse_values = model.history.history["mse"]
mse_values
```

```python
# Predicting the values
Y_pred = model.predict([X])
Y_pred
```

```python
# scatter plot between actual vs predictions
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, y, color='red')
plt.scatter(X, Y_pred)
plt.plot(X, Y_pred, color='green')
plt.ylabel("y and y predictions")
plt.xlabel("epochs")
plt.title("Actual vs predictions")
plt.show()
```

```python
# Not satisfactory, add more EPOCHS
```

```python
# see the error how it is reduced with 5 epochs
import matplotlib.pyplot as plt

plt.scatter(range(1, 21), mse_values)
plt.plot(range(1, 21), mse_values, color='red')
plt.ylabel("Mean squared error")
plt.xlabel("epochs")
plt.show()
```

```python
# Increase EPOCHS as result is not satisfactory
# random state
tf.random.set_seed(42)

# create a model using the sequential API
model= tf.keras.Sequential([tf.keras.layers.Dense(1)])

# compile the model
model.compile(loss=tf.keras.losses.mse,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mse"])

# Fit the model with 100 EPOCHS
model.fit(tf.expand_dims(X, axis = -1), y, epochs = 100)
```

```python
# Saving all the mse values in a variable

mse_values = model.history.history["mse"]
len(mse_values)
```

```python
Y_pred1 = model.predict([X])
Y_pred1
```

```python
# see the error how it is reduced with 100 epochs
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.scatter(range(1, 101), mse_values)
plt.plot(range(1, 101), mse_values, color='red')
plt.ylabel("Mean squared error")
plt.xlabel("epochs")
plt.show()
```

```python
# scatter plot between actual vs predictions
import matplotlib.pyplot as plt

plt.scatter(X, y, color='black')
plt.plot(X, y, color='red')
plt.scatter(X, Y_pred,color='black')
plt.plot(X, Y_pred, color='green')
plt.scatter(X, Y_pred1,color='black')
plt.plot(X, Y_pred1, color='blue')
plt.ylabel("y and y predictions")
plt.xlabel("epochs")
plt.title("Actual vs predictions")
plt.show()
```

```python
# We came close to BEST FIT LINE, Error got reduced between Actual and Predicted
```
