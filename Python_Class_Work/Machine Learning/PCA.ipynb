```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("breast-cancer-wisconsin-data.csv")
df
```

```python
df.info()
```

```python
X = df.iloc[:,2:]
X
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X  = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(df.iloc[:,2:])
SS_X
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["diagnosis"] = LE.fit_transform(df["diagnosis"])
Y = df["diagnosis"]
Y
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(SS_X)
pc_df = pd.DataFrame(pc)
pc_df.head()
```

```python
pc_df.var()
```

```python
# data partition
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(SS_X, Y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
train_acc = []
test_acc = []

for i in range(1,101):
    X_train,X_test,Y_train,Y_test= train_test_split(SS_X,Y,test_size=0.3,random_state=i)
    model.fit(X_train,Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    train_acc.append(accuracy_score(Y_train,Y_pred_train))
    test_acc.append(accuracy_score(Y_test,Y_pred_test))

print("Cross validation Shuffle Split: Training accuracy:",np.round(np.mean(train_acc),2))
print("Cross validation Shuffle Split: Test accuracy:",np.round(np.mean(test_acc),2))
```
