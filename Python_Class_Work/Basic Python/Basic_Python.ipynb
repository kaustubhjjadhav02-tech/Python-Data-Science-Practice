# Logical operators (and, or, not):
1. and: it returns True when all the conditions are found True
2. or: it returns True if either of the conditions are True
3. not: complement(opposite)

```python
username = 'abc'
pwd = 1234
```

```python
username == 'abc' and pwd == 1234
```

```python
username == 'abc' and pwd == 1235
```

```python
username == 'abc' or pwd == 1234
```

```python
username == 'xyz' or pwd == 1234
```

```python
not(username == 'abc')
```

```python
not(username == 'xyz')
```
