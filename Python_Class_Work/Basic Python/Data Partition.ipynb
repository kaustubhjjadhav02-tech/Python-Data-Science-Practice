```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("breast_cancer.csv")
df
```

```python
df.info()
```

```python
X = df.iloc[:,0:9]
X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_sc = sc.fit_transform(X)
X_sc = pd.DataFrame(X_sc)
X_sc.columns = list(X)
X_sc
```

```python
df["Class"] = df["Class"].replace({2:0,4:1})
df["Class"]
```

```python
Y = df["Class"]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_sc, Y, test_size=0.3,random_state = 100)
```

```python
# Model fitting
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# Model Prediction
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
Y_pred_train
```

```python
Y_pred_test
```

```python
# Metrics
from sklearn.metrics import accuracy_score, log_loss
score_train = accuracy_score(Y_train, Y_pred_train)
print("Training Accuracy Score: ",np.round(score_train,3))
error_train = log_loss(Y_train, Y_pred_train)
print("Training Log Loss error: ",np.round(error_train,3))


score_test = accuracy_score(Y_test, Y_pred_test)
print("Testing Accuracy Score: ",np.round(score_test,3))
error_test = log_loss(Y_test, Y_pred_test)
print("Testing Log Loss error: ",np.round(error_test,3))
```

```python
# Training and Testing Data (Data partition)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_sc, Y, test_size=0.3,random_state = 100)

# Model fitting
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)

# Model Prediction
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)

# Metrics
from sklearn.metrics import accuracy_score
score_train = accuracy_score(Y_train, Y_pred_train)
print("Training Accuracy Score: ",np.round(score_train,3))
score_test = accuracy_score(Y_test, Y_pred_test)
print("Testing Accuracy Score: ",np.round(score_test,3))
```

```python
train_acc = []
test_acc = []

for i in range(1,101):
  X_train, X_test, Y_train, Y_test = train_test_split(X_sc, Y, test_size=0.3,random_state = i)
  model.fit(X_train, Y_train)
  Y_pred_train = model.predict(X_train)
  Y_pred_test = model.predict(X_test)
  train_acc.append(accuracy_score(Y_train, Y_pred_train))
  test_acc.append(accuracy_score(Y_test, Y_pred_test))
  score_test = accuracy_score(Y_test, Y_pred_test)

print("Shuffle split cv training score: ",np.round(np.mean(train_acc),3))
print("Shuffle split cv testing score: ",np.round(np.mean(train_acc),3))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5) # by default shuffle = False

train_acc_1 = []
test_acc_1 = []

for train_index, test_index in kf.split(X_sc):
  X_train, X_test, Y_train, Y_test = X_sc.iloc[train_index], X_sc.iloc[test_index], Y[train_index],Y[test_index]
  model.fit(X_train, Y_train)
  Y_pred_train = model.predict(X_train)
  Y_pred_test = model.predict(X_test)
  train_acc_1.append(accuracy_score(Y_train, Y_pred_train))
  test_acc_1.append(accuracy_score(Y_test, Y_pred_test))

print("K Fold cv training score: ",np.round(np.mean(train_acc_1),3))
print("K Fold cv testing score: ",np.round(np.mean(test_acc_1),3))
```
