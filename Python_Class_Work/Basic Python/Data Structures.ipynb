# TUPLE:
1. It is represented by '()'(optional).
2. Insertion order is preserved (Indexing & Slicing).
3. Duplicate values are allowed.
4. Heterogenous elements are allowed.
5. It is IMMUTABLE.
(Read only version of LIST)

```python
# Creation of tuple
# empty tuple
t = ()
print(t)
print(type(t))
```

```python
# single element
t = (10)
print(t)
print(type(t))
```

```python
# single element
t = (10,)
print(t)
print(type(t))
```

```python
# Multiple element
t = 10,10,20,20,30,3.14,"python"
print(t)
print(type(t))
```

```python
# How to access elements
# 1. Indexing
# 2. Slicing
```

```python
# Indexing
t = 10,10,20,20,30,3.14,"python"
print(t[0])
print(t[-1])
```

```python
# Slicing
t = 10,10,20,20,30,3.14,"python"
print(t[::])
print(t[::2])
```

```python
# tuple vs Mutability
t1 = (10,20,20,30,40)
print(t1)
t1[3] = 777
print(t1)
```

```python
# tuple vs Mutability
t1 = [10,20,20,30,4]
print(t1)
t1[3] =777
print(t1)
```

```python
t1 = (10,20,20,30,4)
print(len(t1))
print(t1.count(20))
print(t1.index(20))
print(sorted(t1))
print(sorted(t1,reverse=True))
```

```python
t1 = 10,20,30
t2 = 40,50,60
t3 = t1 + t2
t3
```

```python
t1 = 10,20,30
t2 = t1 * 3
t2
```

```python
l = [1,2,3,5]
print(l)
t = tuple(l)
print(t)
```

```python
t = 10,20,30
print(t)
l = list(t)
print(l)
```

# SET:
1. It is represented by '{}'.
2. Insertion order is not preserved.
3. Duplicate values are not allowed.
4. Heterogenous elements are allowed.
5. It is MUTABLE.

```python
# Creation of Set
# empty set
s = {}
print(s)
print(type(s))
```

```python
s = set()
print(s)
(print(type(s)))
```

```python
# single element
s = {10}
print(s)
print(type(s))
```

```python
# multiple elements
s = {10,10,10,3.14,"python"}
print(s)
```

```python
s = {10,10,10,3.14,"python"}
print(s[0])
```

```python
# set vs mutability
s = {10,20,30,20,10,40,20}
s.add(777)
s
```

```python
# update(iterable)
s = {10,20,30,20,10,40,20}
s.update(777)
s
```

```python
s = {10,20,30}
l = [40,50]
s.update(l)
s
```

```python
s = {10,20,30}
l = [40,50]
s.update(l,range(5))
s
```

```python
s = {10,20,30}
l = [40,50]
s.update(l,range(5),"python")
s
```

```python
t = (10,20)
s = {10,20,30}
s.update(t,"python",range(3))
s
```

```python
s = {10,20,30,15,5,12,"abc"}
print(s.pop())
```

```python
s = {10,20,30,15,5,12,"abc"}
s.remove(70)
s
```

```python
s = {10,20,30,15,5,12,"abc"}
s.discard(70)
s
```

# DICTIONARY:
1. It is represented by '{k:v}'.
2. Insertion order is preserved. 
3. Duplicate values are allowed but there should be UNIQUE keys.
4. Heterogenous elements are allowed.
5. It is MUTABLE.

```python
# Creation of dictionary
# empty dictionary
d = {}
print(d)
print(type(d))
```

Unique : which ooccurs only ones ex. Aadhar card, Pan card etc.
Distinct : count multiple names as 1

- Distinct Values:
Definition: Distinct values refer to all the different values present in a collection, regardless of how many times each value appears. If a value appears multiple times, it is still counted as one distinct value.

- Unique Values:
Definition: Unique values, in a stricter sense, refer to values that appear only once in a collection. If a value appears more than once, it is not considered unique in this context.
