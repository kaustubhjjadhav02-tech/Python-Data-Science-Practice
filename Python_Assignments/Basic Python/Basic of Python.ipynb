Q1. Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
def is_prime(num):
    # Prime numbers are greater than 1
    if num <= 1:
        return False

    # Check divisibility from 2 to sqrt(num)
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False

    return True
    
# Take input from user
number = int(input("Enter a number: "))

if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

Q2. Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random

# Generate two random numbers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask user to calculate the product
print(f"What is {num1} × {num2}?")
user_answer = int(input("Enter your answer: "))

# Correct answer
correct_answer = num1 * num2

# Check if user is correct
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Wrong. The correct answer is {correct_answer}.")
```

Q3. Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
# Python script to print and store squares of either even or odd numbers (100–200)

print("Choose an option:")
print("1. Even numbers")
print("2. Odd numbers")

choice = input("Enter your choice (1 or 2): ").strip()

results = []  # list to store squares

if choice == "1":
    print("\nSquares of even numbers from 100 to 200:")
    results = [num**2 for num in range(100, 201) if num % 2 == 0]
    for num in range(100, 201):
        if num % 2 == 0:
            print(f"{num}^2 = {num**2}")

elif choice == "2":
    print("\nSquares of odd numbers from 100 to 200:")
    results = [num**2 for num in range(100, 201) if num % 2 != 0]
    for num in range(100, 201):
        if num % 2 != 0:
            print(f"{num}^2 = {num**2}")

else:
    print("Invalid choice! Please run again and enter 1 or 2.")

# Show the results list
if results:
    print("\nResults stored in list:")
    print(results)
```

Q4. Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400

```python
# Hospital Billing System

# Charges
room_charge_per_day = 2000
consultation_fee = 1500
lab_test_charge = 300
discount_threshold = 10000
discount_rate = 0.10

# Patient details
days_stayed = 4
lab_tests = 3
medicine_charges = 2400

# Bill calculation
room_total = days_stayed * room_charge_per_day
lab_total = lab_tests * lab_test_charge
subtotal = room_total + consultation_fee + lab_total + medicine_charges

# Apply discount if applicable
if subtotal > discount_threshold:
    discount = subtotal * discount_rate
else:
    discount = 0

final_bill = subtotal - discount

# Display bill
print("------ Hospital Bill ------")
print(f"Room charges ({days_stayed} days): ₹{room_total}")
print(f"Doctor consultation fee: ₹{consultation_fee}")
print(f"Lab test charges ({lab_tests} tests): ₹{lab_total}")
print(f"Medicine charges: ₹{medicine_charges}")
print(f"Subtotal: ₹{subtotal}")
print(f"Discount: ₹{discount}")
print(f"Final Bill: ₹{final_bill}")
```
