Q1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# List Operations in Python

# 1. Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original List:", numbers)

# 2. Append - add a single element at the end
numbers.append(60)
print("After append(60):", numbers)

# 3. Extend - add multiple elements at the end
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# 4. Insert - add an element at a specific index
numbers.insert(2, 25)  # insert 25 at index 2
print("After insert(2, 25):", numbers)

# 5. Remove - remove first occurrence of a value
numbers.remove(40)  # removes 40
print("After remove(40):", numbers)

# 6. Pop - remove element at given index (or last element if no index)
popped_value = numbers.pop(3)  # remove element at index 3
print(f"After pop(3): {numbers}, Popped value: {popped_value}")

# 7. Index - find the index of first occurrence of a value
index_value = numbers.index(25)
print("Index of 25:", index_value)

# 8. Count - count occurrences of a value
count_value = numbers.count(20)
print("Count of 20:", count_value)

# 9. Sort - sort list in ascending order
numbers.sort()
print("After sort():", numbers)

# 10. Reverse - reverse the list order
numbers.reverse()
print("After reverse():", numbers)

# 11. Clear - remove all elements from list
numbers.clear()
print("After clear():", numbers)
```

Q2. Create a tuple that stores 3 student names. Try changing the second name in the tuple. What happens? Explain why.

```python
# Creating a tuple with 3 student names
students = ("Amit", "Priya", "Rohan")
print("Original tuple:", students)

# Trying to change the 2nd element (index 1)
try:
    students[1] = "Neha"  # Attempt to modify tuple
except TypeError as e:
    print("Error:", e)
```

Explanation:
1. Tuples in Python are immutable, meaning once created, their elements cannot be changed (no assignment, removal, or insertion).
2. That’s why when you try to replace "Priya" with "Neha", Python raises a TypeError.
3. If you need a collection where elements can be changed, you should use a list instead.

## Q6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'

```python
# Take marks as input
marks = int(input("Enter marks: "))

# Check grade conditions
if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 50:
    grade = "C"
else:
    grade = "Fail"

# Print result
print(f"Marks: {marks}, Grade: {grade}")
```

Q7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
# Take input from user
num = int(input("Enter a number: "))

# Check conditions
if num > 0:
    if num % 2 == 0:
        print(f"{num} is positive and even.")
    else:
        print(f"{num} is positive but odd.")
elif num == 0:
    print("The number is zero.")
else:
    print(f"{num} is negative.")
```

Q11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows

```python
import pandas as pd
import numpy as np

# 1. Create a DataFrame manually with 10 rows
data = {
    'Name': ['Amit', 'Priya', 'Rohan', 'Neha', 'Anil', 'Priya', 'Kiran', 'Rohan', 'Sita', 'Amit'],
    'Age': [25, 30, 28, 22, 35, 30, 27, 28, 26, 25],
    'City': ['Mumbai', 'Delhi', 'Pune', 'Mumbai', 'Delhi', 'Delhi', 'Pune', 'Pune', 'Mumbai', 'Mumbai'],
    'Salary': [50000, 60000, np.nan, 45000, 70000, 60000, 48000, np.nan, 52000, 50000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# 2. Use .info() and .describe()
print("\nDataFrame Info:")
print(df.info())

print("\nDataFrame Describe:")
print(df.describe())

# 3. Select 'Name' and 'City' columns
print("\nSelected Columns (Name and City):")
print(df[['Name', 'City']])

# 4. Drop 'City' column
df_dropped = df.drop('City', axis=1)
print("\nDataFrame after dropping 'City' column:\n", df_dropped)

# 5. Fill missing values in 'Salary' with the mean
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)
print("\nDataFrame after filling missing Salary values:\n", df)

# 6. Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:\n", df_no_duplicates)
```
